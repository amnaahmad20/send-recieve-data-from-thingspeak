#send dummy data using random function to thinkspeak

```python
import requests
import random
import time

THINGSPEAK_WRITE_API_KEY = ''  # Your ThingSpeak Write API Key
THINGSPEAK_URL = 'https://api.thingspeak.com/update'

def send_data_to_thingspeak(temp, gas, rain, wind_speed):
    data = {
        'api_key': THINGSPEAK_WRITE_API_KEY,
        'field1': temp,
        'field2': gas,
        'field3': rain,
        'field4': wind_speed
    }
    response = requests.get(THINGSPEAK_URL, params=data)
    print(f"Data sent to ThingSpeak. Response Code: {response.status_code}")

def generate_dummy_data():
    # Simulate temperature (in degrees Celsius)
    temp = random.uniform(15, 35)  # Typical range between 15°C and 35°C

    # Simulate gas sensor value (arbitrary units)
    gas = random.randint(100, 400)  # Assuming a scale of 100 to 400

    # Simulate rainfall (in mm)
    rain = random.uniform(0, 20)  # Rainfall range between 0 mm and 20 mm

    # Simulate wind speed (in km/h)
    wind_speed = random.uniform(0, 100)  # Wind speeds between 0 km/h and 100 km/h

    return temp, gas, rain, wind_speed

# Sending simulated data to ThingSpeak at regular intervals
while True:
    simulated_temp, simulated_gas, simulated_rain, simulated_wind_speed = generate_dummy_data()
    send_data_to_thingspeak(simulated_temp, simulated_gas, simulated_rain, simulated_wind_speed)
    time.sleep(5)  # Pause for 5 seconds before sending the next set of data
    #you can change the time according to your requirements
```

#sending test data from a CSV to Thinkspeak

```python
import requests
import time
import pandas as pd

THINGSPEAK_WRITE_API_KEY = '' #Replace with your channel WRITE API KEY
THINGSPEAK_URL = 'https://api.thingspeak.com/update'

def send_data_to_thingspeak(temp, gas, rain, wind_speed):
    data = {
        'api_key': THINGSPEAK_WRITE_API_KEY,
        'field1': temp,
        'field2': gas,
        'field3': rain,
        'field4': wind_speed
    }
    response = requests.get(THINGSPEAK_URL, params=data)
    print(f"Data sent to ThingSpeak. Response Code: {response.status_code}")

# Load data from CSV file
df = pd.read_csv('/content/data-test-isb.csv')

#  CSV has columns named 'temperature', 'gas', 'rain', 'wind_speed'
for index, row in df.iterrows():
    send_data_to_thingspeak(row['temperature'], row['gas'], row['rain'], row['wind_speed'])
    time.sleep(15)  # ThingSpeak has a rate limit for data upload
```
